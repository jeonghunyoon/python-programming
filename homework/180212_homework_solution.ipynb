# Problem 1

> 1부터 10까지의 범위내에서, 랜덤하게 정수 10개를 불러와서, 리스트 A, B를 생성

```python
import random
import numpy as np
from IPython.display import display

""""""
# 리스트 A의 Seed : 43
random.seed(43)
list_A = [random.randrange(1, 11) for _ in xrange(10)]

print "list_A : "
display(list_A)

# 리스트 B의 Seed : 34
random.seed(34)
list_B = [random.randrange(1, 11) for _ in xrange(10)]

print "list_B : "
display(list_B)
```

> A, B를 unit vector로 만드는 함수 생성 후, 단위 벡터를 생성

```python
# 함수 정의
def get_unit_vector(x):
    """벡터의 길이를 구한 후에, 
    각 원소를 벡터의 길이로 나누어준다."""
    len_of_x = np.sqrt(sum([x_i*x_i for x_i in x]))
    return [(1/len_of_x) * x_i for x_i in x]

unit_A = get_unit_vector(list_A)
unit_B = get_unit_vector(list_B)

print "unit_A : "
display(unit_A)

print "unit_B : "
display(unit_B)
```

> 벡터의 거리 구하기

```python
# 거리 구하는 벡터를 만들어도 되지만, list comprehension을 사용하면 된다.
np.sqrt(sum([(x_i - y_i)**2 for x_i, y_i in zip(unit_A, unit_B)]))
```

# Problem 3

```python
class Calculator:
    
    def __init__(self, x):
        # 리스트를 input으로 받는다. 이것이 보장이 안되면, sum, len 함수 사용시 에러!
        if type(x) != list:
            raise Exception("리스트를 입력값으로만 받습니다!")
        self.x = x
        
    # 합계를 구하는 메서드를 정의한다.
    def sum(self):
        return sum(self.x)
    
    # 평균을 구하는 메서드를 정의한다.
    def avg(self):
        return float(sum(self.x)) / len(self.x)

cal1 = Calculator([1, 2, 3, 4, 5])
display(cal1.sum())
display(cal1.avg())

cal2 = Calculator([6, 7, 8, 9, 10])
display(cal2.sum())
display(cal2.avg())
```

# Problem 5

> Gradient를 구한다.

```python
def get_gradient(v):
    """편미분 함수를 구한다."""
    return [4 * (v_i ** 3) for v_i in v]
```

> Step을 구한다.

```python
def step(v, gradient, step_size):
    """현재 위치에서 step_size와 gradient 이용하여 이동한다."""
    return [v_i + step_size * direction_i 
            for v_i, direction_i in zip(v, gradient)]
```

> 시작점은 랜덤하게 선택해보자

```python
import random
v = [random.randint(-10, 10) for _ in range(3)]

v
```

> 학습률을 설정하자

```python
alpha = -0.01
```

> 알고리즘을 멈출 조건을 설정하자.

```python
import math
def distance(v, w):
    return math.sqrt(sum([(v_i - w_i)**2 for v_i, w_i in zip(v, w)]))
```

```python
while True:
    gradient = get_gradient(v)
    next_v = step(v, gradient, alpha)
    print next_v
    if distance(next_v, v) < 0.000001:
        break
    v = next_v
```

# Problem 6

```python
def compress(li):
    # 결과값으로 반환될 문자열이다. 즉 압축된 문자열이다.
    compressed_str = ""
    
    # 현재 문자와 비교할, 이전의 문자이다.
    # 초기값은 문자열의 첫번째 문자로 셋팅한다.
    last_ch = li[0]
    
    # 문자의 발생 횟수
    # 초기값은 문자열의 첫번째 문자를 카운팅했으니 1로 한다.
    cnt = 1
        
    # 문자열의 2번째 문자부터 체크함에 주의하라.
    for ch in li[1:]:
        if ch == last_ch:
            cnt+=1
        else:
            compressed_str = compressed_str + str(last_ch) + str(cnt)
            cnt = 1
            last_ch = ch
            
    # 마지막에 scan되는 문자들을 압축해준다.
    compressed_str = compressed_str + str(last_ch) + str(cnt)
            
    return compressed_str
            
compress("aaabbcccccca")
```

# Problem 7

> 입력된 문자열을 분석하여 각각의 문자의 빈도수가 1 이상이면 False를, 아니면 True를 반환하는 함수이다.

```python
def check_freq(numbers):
    # 길이가 10이 아니면 False를 리턴한다.
    if len(numbers) is not 10 :
        return False
    
    # dictionary에서 key 값이 존재한다면 숫자가 이미 나온것이다.
    nums = {}
    for number in numbers:
        if number in nums:
            return False
        else:
            nums[number] = 1
    return True
```

> 입력된 문자열 리스트를 split 하여, 각각의 원소에 대하여 check_freq 함수를 호출

```python
def check_number(li):
    # 숫자가 1개 이상의 리스트로 입력된 경우를 대비
    tokens = li.split(" ")
    
    for token in tokens:
        print check_freq(token)
```

# Problem 8

> 친구 관계를 입력으로 받아서 인접리스트(adjacency list)를 반환

```python
def get_adj_list(relationships):    
    # 친구 관계를 담고 있는 dictionary
    adj_list = {}
    
    for i, j in relationships:
        # i의 친구 관계를 체크한다.
        if i in adj_list:
            adj_list[i].add(j)
        else:
            friends = set([j])
            adj_list[i] = friends
            
        # j의 친구 관계를 체크한다.
        if j in adj_list:
            adj_list[j].add(i)
        else:
            friends = set([i])
            adj_list[j] = friends

    return adj_list
```

> 사람을 입력받으면 그 사람의 친구 리스트를 출력한다.

```python
def get_friends(i, friends):
    adj_list = get_adj_list(friends)
    return list(adj_list[i])
```
