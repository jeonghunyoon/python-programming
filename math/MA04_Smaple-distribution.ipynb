# Chi square distribution(카이제곱분포)

```python
from scipy.stats import chi2, norm, f, t
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = (10,5)
```

```python
# Pdf of chi2
x1 = np.linspace(0.01, 10, 100)

for df in [1, 2, 3, 4, 5, 6, 9]:
    rv = chi2(df = df)
    y1 = rv.pdf(x1)
    plt.plot(x1, y1, label=("chi2 (degree of freedom = %d)" %df))

plt.xlim(0, 10.1)
plt.ylim(0, 0.6)
plt.legend()
plt.show()
```

```python
# Random samples from normal distribution
rand = np.random.RandomState(0)

rv = norm()

# 1 by 1000 normal variable matrix
x1 = rv.rvs((1, 1000))
# 3 by 1000 normal variable matrix
x2 = rv.rvs((3, 1000))
# 3 by 1000 normal variable matrix
x3 = rv.rvs((9, 1000))

plt.subplot(1, 3, 1)
sns.distplot((x1 ** 2).sum(axis=0), kde = False)
plt.xlim(-1, 10)
plt.ylim(0, 500)
plt.title("number of sample = 1")

plt.subplot(1, 3, 2)
sns.distplot((x2 ** 2).sum(axis=0), kde = False)
plt.xlim(-1, 10)
plt.ylim(0, 500)
plt.title("number of sample = 3")

plt.subplot(1, 3, 3)
sns.distplot((x3 ** 2).sum(axis=0), kde = False)
plt.xlim(-1, 10)
plt.ylim(0, 500)
plt.title("number of sample = 9")

plt.show()
```

# F distribution (F 분포)

```python
# Pdf of f distribution
x1 = np.linspace(0.03, 3, 1000)
df = [(1, 1), (1, 5), (5, 5), (10, 10)]
for i, j in df:
    y1 = f(i, j).pdf(x1)
    plt.plot(x1, y1, label=("F(%i, %i)" %(i, j)))

plt.legend()
plt.show()
```

```python
# Random samples from normal distribution

rv = norm()

x2 = rv.rvs((5, 10000))
x3 = rv.rvs((1, 10000))

x4 = rv.rvs((10, 10000))
x5 = rv.rvs((10, 10000))

plt.subplot(1,2,1)
sns.distplot((x3 ** 2).sum(axis=0) / (x2 ** 2).sum(axis=0), bins=200, kde = False)
plt.xlim(-0.1, 3)
plt.title("N1=1, N2=5")

plt.subplot(1,2,2)
sns.distplot((x5 ** 2).sum(axis=0) / (x4 ** 2).sum(axis=0), bins=200, kde = False)
plt.xlim(-0.1, 3)
plt.title("N1=10, N2=10")

plt.show()
```

# T distribution (t 분포)

```python
# Pdf of T distribution
x1 = np.linspace(-4, 4, 100)
for df in [1, 5, 10, 20]:
    rv = t(df = df)
    y1 = rv.pdf(x1)
    plt.plot(x1, y1, label=("student-t (degree of freedom=%d)" %df))

plt.plot(x1, norm.pdf(x1), label="Normal distribution", lw = 5, alpha=0.5)
plt.legend()
plt.show()
```
