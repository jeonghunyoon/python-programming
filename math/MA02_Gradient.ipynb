# 경사하강법

$f(x,y,z)=x^2+y^2+z^2$ 의 최소값을 경사하강법을 이용하여 구하라.

### 1. gradient를 구한다.

```python
def get_gradient(v):
    """편미분 함수를 구한다."""
    return [2 * v_i for v_i in v]
```

### 2. step을 구한다.

```python
def step(v, gradient, step_size):
    """현재 위치에서 step_size와 gradient 이용하여 이동한다."""
    return [v_i + step_size * direction_i 
            for v_i, direction_i in zip(v, gradient)]
```

### 3. 시작점은 랜덤하게 선택해보자.

```python
import random
v = [random.randint(-10, 10) for _ in range(3)]

v
```

### 4. alpha (학습률)를 설정하자.

```python
alpha = -0.01
```

### 5. 알고리즘을 멈출 조건을 설정하자.

```python
import math
def distance(v, w):
    return math.sqrt(sum([(v_i - w_i)**2 for v_i, w_i in zip(v, w)]))
```

```python
while True:
    gradient = get_gradient(v)
    next_v = step(v, gradient, alpha)
    print next_v
    if distance(next_v, v) < 0.0000001:
        break
    v = next_v
```

### 퀴즈

<U>**Rosenbrock function**</U>

$f(x, y) = (1-x)^2+100(y-x^2)^2$ 의 최소값을 경사하강법을 (또는 최대값을 경사상승법을) 사용하여 구하라.

# 회귀분석

### 1. gradient를 구해보자.

```python
def get_gradient_reg(x, y, w):
    """편미분 함수를 구한다."""
    return [ 0.1 * sum([-2 * (y_i - w[0] - w[1]*x_i) 
                  for x_i, y_i in zip(x, y)]),
            0.1 * sum([-2 * x_i *(y_i - w[0] - w[1]*x_i) 
                 for x_i, y_i in zip(x, y)]) ]
```
